```python
state_abbrev = {'California': 'CA', 
                'New York': 'NY', 
                'Illinois': 'IL', 
                'Florida': 'FL', 
                'Nebraska':'NE'}

def abbreviate_state(df):
    if df['state'] in state_abbrev.keys():
        abbrev = state_abbrev[df['state']]
        return abbrev
    else:
        return df['state']

patients_cdf['state'] = patients_cdf.apply(abbreviate_state, axis=1)
patients_cdf.head()
```

Treatments Table

```python
treatment_df.head()
```

```python
treatment_cdf = treatment_df.copy()
```

```python
treatment_cdf[treatment_cdf.duplicated()]
treatment_cdf = treatment_cdf.drop_duplicates()
treatment_cdf = treatment_cdf.reset_index(drop=True)
```

```python
treatment_cdf.given_name = treatment_cdf.given_name.str.title()
treatment_cdf.surname = treatment_cdf.surname.str.title()
treatment_cdf.head()
```

```python
treatment_cdf.hba1c_change = treatment_cdf.hba1c_start - treatment_cdf.hba1c_end
```

```python
treatment_cdf.head()
```

```python
treatment_cdf = pd.melt(treatment_cdf,id_vars=['given_name','surname','hba1c_start','hba1c_end','hba1c_change'],var_name='treatments' ,
                        value_name='dose')
```

```python
treatment_cdf.head()
```

```python
treatment_cdf = treatment_cdf[treatment_cdf.dose != '-']
```

```python
treatment_cdf = treatment_cdf.reset_index(drop=True)
treatment_cdf.head()
```

```python
treatment_cdf['start_dose'] ,treatment_cdf['end_dose'] = treatment_cdf.dose.str.split(' - ', 1).str
treatment_cdf = treatment_cdf.drop('dose',axis=1)
treatment_cdf.head()
```

```python
treatment_cdf['end_dose'] = treatment_cdf.end_dose.str.strip('u').astype(int)
treatment_cdf['start_dose'] = treatment_cdf.start_dose.str.strip('u').astype(int)
```

```python
treatment_cdf.head()
```

```python
treatment_cdf['change_dose'] = treatment_cdf['end_dose'] - treatment_cdf['start_dose']
```

```python
treatment_cdf.head()
```
